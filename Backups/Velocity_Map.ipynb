```python
def w_v(wls, restwl):

    """
    This function converts wavelengths to velocity in km/s.
    """
    Vsys = 243
    c = 3.0*10**5

    vels = (wls*c / restwl) - c

    v_rot = vels - Vsys
    v_obs = v_rot * sin(78) # accounts for disk inclination
    
    return v_obs

'''NIIa = 6548.05
Halpha = 6562.801
NIIb = 6583.45
SIIa = 6716.44
SIIb = 6730.82

Vsys = 243'''
```

```python
# creating velocity columns for each emission + component
df['Vel1'] = w_v(df['Wvl1'], 6548.05)
df['Vel2'] = w_v(df['Wvl2'], 6548.05)
df['Vel3'] = w_v(df['Wvl3'], 6562.801)
df['Vel4'] = w_v(df['Wvl4'], 6562.801)
df['Vel5'] = w_v(df['Wvl5'], 6583.45)
df['Vel6'] = w_v(df['Wvl6'], 6583.45)
df['Vel7'] = w_v(df['Wvl7'], 6716.44)
df['Vel8'] = w_v(df['Wvl8'], 6716.44)
df['Vel9'] = w_v(df['Wvl9'], 6730.82)
df['Vel10'] = w_v(df['Wvl10'], 6730.82)
df
```

```python
def velmap(arr, title):
    # creates velocity map with fixed colorbar
    fig, ax = plt.subplots()
    fig = plt.imshow(arr, cmap = "bwr")
    cl = ax.imshow(arr, norm=colors.CenteredNorm(0), cmap='bwr')
    plt.colorbar(cl,label = 'Velocity km/s')
    ax.set_title(title)
    plt.xlabel('R.A.')
    plt.ylabel('Dec.')
    plt.ylim(0, 34)
    return plt.show()
    
velmap(arr, "Velocity Map: 200-234 x 200-234")
## practice --> shows Vel1 for 200-234 x 200-234
```
